```python
import numpy as np
import pandas as pd
from sklearn.preprocessing import OneHotEncoder
```

```python
dataset = pd.read_csv('height_weight.csv')
```

```python
X = dataset.iloc[:,:-1]
Y = dataset.iloc[:, -1]
```

```python
X = np.array(X)
Y = np.array(Y)
print(X)
print(Y)
```

```python
encoder = OneHotEncoder()
Y_encoded = encoder.fit_transform(Y.reshape(-1, 1)).toarray()
```

```python
predicted_value = [[161, 61]]
mean_X = np.mean(X, axis=0)
std_X = np.std(X, axis=0)
print(mean_X,"===================", std_X)
standardized_X = (X - mean_X) / std_X
standardized_predicted_value = (predicted_value - mean_X) / std_X
print("Standard X Value : " , standardized_X)
print("Standard Predicted Value : " , standardized_predicted_value)
```

```python
def knn(X, Y, query_point, k):
    # Calculate distances between the query point and all points in the dataset
    distances = np.linalg.norm(X - query_point, axis=1)
    # Sort the distances and get the indices of the sorted distances
    for i, distance in enumerate(distances):
      print(f"Point {X[i]}: {distance}")
    sorted_indices = np.argsort(distances)
    # Select the indices of the k nearest neighbors
    k_nearest_indices = sorted_indices[:k]
    k_nearest_labels = Y[k_nearest_indices]
    print(k_nearest_labels,"k_nearest_labels")
    prediction = np.argmax(np.sum(k_nearest_labels, axis=0))
    print(prediction,"=============prediction")
    print("Prediction for the predicted value:", encoder.categories_[0][prediction])
    # Determine the labels of the k nearest neighbors
    return k_nearest_indices

# Find the closest neighbors
k = 5  # Number of nearest neighbors to find
nearest_indices = knn(standardized_X, Y_encoded, standardized_predicted_value, k)

print(f"The {k} closest neighbors to the query point are:")
for index in nearest_indices:
    print(X[index])
```
