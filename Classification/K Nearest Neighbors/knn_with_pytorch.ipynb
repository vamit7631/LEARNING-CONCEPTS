```python
import torch
import pandas as pd
```

```python
dataset = pd.read_csv('Social_Network_Ads.csv')
```

```python
X = dataset.iloc[:, :-1]
Y = dataset.iloc[:, -1]
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size = 0.25, random_state=0)
```

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
X_train = sc.fit_transform(X_train)
X_test = sc.transform(X_test)
```

```python
X_train_tensor = torch.tensor(X_train, dtype = torch.float32)
X_test_tensor = torch.tensor(X_test, dtype = torch.float32)
Y_train_tensor = torch.tensor(Y_train.values, dtype = torch.float32)
Y_test_tensor = torch.tensor(Y_test.values, dtype = torch.float32)
```

```python
def knn_predict(X_train, X_test, Y_train, k):
    distances = torch.cdist(X_test, X_train)
    _, indices = distances.topk(k, dim=1, largest=False)
    knn_labels = Y_train[indices]
    predictions = knn_labels.mode(dim=1).values.flatten()
    return predictions
```

```python
Y_pred = knn_predict(X_train_tensor,X_test_tensor, Y_train_tensor, k=5)
```

```python
from sklearn.metrics import accuracy_score, confusion_matrix
accuracy = accuracy_score(Y_test_tensor.numpy(), Y_pred.numpy())
confusiondata = confusion_matrix(Y_test_tensor.numpy(), Y_pred.numpy())
print(f"Accuracy: {accuracy}")
print(f"Confusion Matrix \n: {confusiondata}")
```
